# 01 · Análisis exploratorio

Carga y descripción inicial de las respuestas exportadas de Tally.

**Prerequisito:** exportar respuestas desde Tally UI a `data/raw/` o correr `analysis/scripts/fetch_tally.py`.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv

load_dotenv()

sns.set_theme(style='whitegrid', palette='deep')
plt.rcParams['figure.dpi'] = 150
```

## Carga de datos

```python
# Ajusta el path si usas CSV exportado manualmente desde Tally
raw_files = sorted(Path('../../data/raw').glob('*.csv'))
if not raw_files:
    raw_files = sorted(Path('../../data/raw').glob('*.json'))

print(f'Archivos encontrados: {[f.name for f in raw_files]}')

# Cargar el más reciente
latest = raw_files[-1]
if latest.suffix == '.csv':
    df = pd.read_csv(latest)
else:
    import json
    subs = json.loads(latest.read_text())
    df = pd.json_normalize(subs)

print(f'n = {len(df)} respuestas')
df.head()
```
